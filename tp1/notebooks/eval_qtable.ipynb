```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from src.main import BlackjackQTable
```

```python
qtable = BlackjackQTable.load(path='../outputs/n_episodes_50000/20231004T210835/episodes_50000__BlackjackQTable.pickle')
```

```python
qtable.updates
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 7), sharex='all')

sns.heatmap(qtable.table_[:, :, 0, 1], ax=ax[0])
ax[0].set_title('HIT action value by state')

sns.heatmap(qtable.table_[:, :, 0, 0], ax=ax[1])
ax[1].set_title('STICK action value by state')

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 7), sharex='all')

arr = qtable.ocurrences_[:, :, 0, 1]
sns.heatmap(arr / arr.sum(), ax=ax[0])
ax[0].set_title('Number of HIT ocurrences by state')

arr = qtable.ocurrences_[:, :, 0, 0]
sns.heatmap(arr / arr.sum(), ax=ax[1])
ax[1].set_title('Number of STICK ocurrences by state')

plt.show()
```
