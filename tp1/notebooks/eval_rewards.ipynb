```python
import polars as pl
import matplotlib.pyplot as plt
```

```python
rewards = pl.read_csv('../outputs/n_episodes_50000/20231008T160348/blackjack__rewards.csv', has_header=False)
```

```python
window_size = int(rewards.shape[0] * 0.1)
window_size
```

```python
plt.figure(figsize=(12, 3))
ax = plt.gca()

(
    rewards
    .to_series()
    .rolling_mean(window_size=window_size)
    .to_pandas()
    .plot(ax=ax)
)
```

```python
plt.figure(figsize=(12, 3))
ax = plt.gca()

(
    rewards
    .to_series()
    .rolling_sum(window_size=window_size)
    .to_pandas()
    .plot(ax=ax)
)
```
