```python
import polars as pl
import matplotlib.pyplot as plt
```

```python
rewards = pl.read_csv('../src/episodes_1000__rewards.csv', has_header=False)
```

```python
(
    rewards
    .mean(axis=1)
    .to_pandas()
    .plot(figsize=(12, 3))
)
```

```python
(
    rewards
    .sum(axis=1)
    .to_pandas()
    .plot(figsize=(12, 3))
)
```
