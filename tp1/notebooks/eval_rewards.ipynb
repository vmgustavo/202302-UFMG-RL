```python
import polars as pl
import matplotlib.pyplot as plt
```

```python
rewards = pl.read_csv('../src/episodes_1000__rewards.csv', has_header=False)
```

```python
plt.figure(figsize=(12, 3))
ax = plt.gca()

(
    rewards
    .sum(axis=1)
    .rolling_mean(window_size=32)
    .to_pandas()
    .plot(ax=ax)
)

(
    rewards
    .sum(axis=1)
    .to_pandas()
    .plot(ax=ax, alpha=.3)
)
```

```python
plt.figure(figsize=(12, 3))
ax = plt.gca()

(
    rewards
    .mean(axis=1)
    .rolling_mean(window_size=32)
    .to_pandas()
    .plot(ax=ax)
)

(
    rewards
    .mean(axis=1)
    .to_pandas()
    .plot(ax=ax, alpha=.3)
)
```
