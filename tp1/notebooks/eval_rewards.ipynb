```python
import polars as pl
import matplotlib.pyplot as plt
```

```python
rewards = pl.read_csv('../outputs/n_episodes_10000/20231004T210317/episodes_10000__rewards.csv', has_header=False)
```

```python
window_size = 256
```

```python
plt.figure(figsize=(12, 3))
ax = plt.gca()

(
    rewards
    .to_series()
    .rolling_mean(window_size=window_size)
    .to_pandas()
    .plot(ax=ax)
)
```

```python
plt.figure(figsize=(12, 3))
ax = plt.gca()

(
    rewards
    .to_series()
    .rolling_sum(window_size=window_size)
    .to_pandas()
    .plot(ax=ax)
)
```
